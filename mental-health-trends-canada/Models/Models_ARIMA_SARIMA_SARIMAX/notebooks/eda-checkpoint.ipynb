### Trend for Independent Variables

```python
plt.figure(figsize=(16, 12))

# ✅ Loop through all independent variables
independent_vars = [
    'Quarterly Avg Temp (°C)',
    'Quarterly Sum Precip (mm)',
    'CPI Average',
    'CPI value Cumulative',
    'CPI Compounded',
    'Debt_to_Asset_Ratio',
    'Mortgage_Debt_to_Real_Estate_Ratio',
    'Depression',
    'Anxiety',
    'Therapy',
    'Stress',
    'Unemployment Rate'
]

# Plot each independent variable
for var in independent_vars:
    sns.lineplot(data=aggregated_df, x=aggregated_df.index, y=var, label=var, errorbar='sd')
    
    # ✅ Add shaded confidence interval manually using fill_between()
    plt.fill_between(aggregated_df.index,
                     aggregated_df[var] - std[var],
                     aggregated_df[var] + std[var],
                     alpha=0.2)

# Format x-axis to "2022-04"
plt.xticks(aggregated_df.index, aggregated_df.index.strftime('%Y-%m'), rotation=45)

# Formatting
plt.title('Independent Variables Over Time (Aggregated)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside
plt.show()
```

## Check Seasonality

```python
from statsmodels.tsa.seasonal import seasonal_decompose

for var in independent_vars:
    result = seasonal_decompose(aggregated_df[var], model='additive', period=4)  # Quarterly data → period=4

    plt.figure(figsize=(12, 8))
    
    # Original data
    plt.subplot(4, 1, 1)
    plt.plot(aggregated_df[var], label='Original', color='blue')
    plt.legend()
    
    # Trend
    plt.subplot(4, 1, 2)
    plt.plot(result.trend, label='Trend', color='green')
    plt.legend()
    
    # Seasonal
    plt.subplot(4, 1, 3)
    plt.plot(result.seasonal, label='Seasonal', color='orange')
    plt.legend()
    
    # Residual (Noise)
    plt.subplot(4, 1, 4)
    plt.plot(result.resid, label='Residual', color='red')
    plt.legend()
    
    plt.suptitle(f'Seasonal Decomposition of {var}')
    plt.tight_layout()
    plt.show()
```

```python
from scipy.signal import periodogram

for var in independent_vars:
    freq, power = periodogram(aggregated_df[var].dropna())
    plt.figure(figsize=(10, 5))
    plt.plot(freq, power)
    plt.title(f'Periodogram of {var}')
    plt.xlabel('Frequency')
    plt.ylabel('Power')
    plt.show()

```

```python
### Kruskal-Wallis Test (Quarterly Seasonality):
```

```python
from scipy.stats import kruskal

for var in independent_vars:
    # Create groups by quarter
    groups = [aggregated_df[var][aggregated_df.index.quarter == q].dropna() for q in range(1, 5)]
    
    # Perform Kruskal-Wallis Test
    stat, p = kruskal(*groups)
    
    print(f'{var} Seasonality Test:')
    print(f'Kruskal-Wallis Statistic: {stat:.2f}, p-value: {p:.5f}')
    if p < 0.05:
        print("✅ Seasonality Detected\n")
    else:
        print("❌ No Significant Seasonality\n")
```

## Correlation Matrix

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Compute correlation matrix
corr_matrix = aggregated_df[independent_vars].corr()

# Create a mask to highlight absolute values > 0.5
high_corr_mask = corr_matrix.abs() > 0.5

plt.figure(figsize=(12, 10))

# Create an upper triangle mask to hide duplicate values
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Plot the heatmap
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5,
    mask=mask,
    cbar=True,
    center=0,
    vmax=1,
    vmin=-1
)

# Highlight values with |correlation| > 0.5 using a bright green border
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if high_corr_mask.iloc[i, j]:
            plt.gca().add_patch(plt.Rectangle((j, i), 1, 1, fill=False, edgecolor='lime', lw=2))

# Add title and show the plot
plt.title('Correlation Matrix (Independent Variables)')
plt.show()
```
